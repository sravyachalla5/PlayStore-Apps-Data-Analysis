## Cleaning Data
* Dataset can contain missing data, numerical string value, various cues. If we can clean them, we can make easy our        analysis.

### Playstore_apps Data

```python
# Check duplicated rows
app_dub = apps_df[apps_df.duplicated()]
print(f' There are {len(app_dub)} duplicated rows in app dataset')

# Remove duplicated rows by keeping first row
apps_df = apps_df.drop_duplicates(keep = 'first')
print(f' There are {len(apps_df)} unique rows in app dataset after removing duplicated rows')
```

```python
# Category
apps_df['Category'].unique()
```

'1.9' is wrong value. So I have to see in which rows '1.9' corresponds to.

```python
apps_df[apps_df['Category'] == '1.9']
```

we can drop this perticular row

```python
apps_df = apps_df.drop(index = 10472,axis=0)
```

```python
# Ratings
apps_df['Rating'].unique()
```

```python
# 1.	As the rating can be a float value change the null values of the rating column to 0
```

```python
apps_df['Rating'].fillna(0,inplace=True)
```

```python
apps_df['Rating'].unique()
```

```python
#Reviews
apps_df['Reviews'].unique()
```

```python
#Size
apps_df['Size'].unique()
```

```python
#Installs
apps_df['Installs'].unique()
```

```python
# Price
apps_df['Price'].unique()
```

```python
#Last Updated
apps_df['Last Updated'].unique()
```

```python
apps_df['Last Updated'] = pd.to_datetime(apps_df['Last Updated'])
apps_df['Last Updated']
```

```python
apps_df['App'].value_counts()
```

As we can see from the above mentioned data, some of the apps are having multiple rows. Let's check out if there data is identical or not

```python
apps_df[apps_df['App'] == 'ROBLOX']
```

As we can see from the above dataframe, ROBLOX app is having identical rows with difference in number of reviews. It may have happened that for the same app, the data has been scraped in different points of time. So we have kept row of an app with maximum number of reviews, assuming it to be the latest one.

```python
apps_df = apps_df.sort_values(by = 'Reviews',ascending=False)
```

```python
apps_df.shape
```

```python
apps_df.drop_duplicates(subset=['App', 'Category', 'Rating', 'Size', 'Installs', 'Type',
   'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver',
   'Android Ver'],keep='first',inplace=True)
```

```python
apps_df.shape
```

```python
apps_df = apps_df.sample(frac = 1).reset_index(drop=True)
```

```python
# Check null values
print('Before cleaning null values, number of null values in each columns :')
print(apps_df.isnull().sum(),'\n')

# Drop rows for remaining rows contain null
apps_df.dropna(inplace = True)

print('After cleaning, number of null values in each columns :')
print(apps_df.isnull().sum())
```
